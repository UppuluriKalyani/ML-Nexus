Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Analysis

```python
# loading the data from csv file to a Pandas DataFrame
parkinsons_data = pd.read_csv('/content/PD_data.csv')
```

```python
# Convert object type columns to numeric before calculating the mean
# First identify the numeric columns using select_dtypes.
numeric_cols = parkinsons_data.select_dtypes(include=['number']).columns

# Next filter the dataframe to only include numeric features using those columns
# this will avoid the TypeError when mean is called
parkinsons_data.loc[:, numeric_cols].groupby('status').mean()
```

Data Pre-Processing

Separating the features & Target

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```
